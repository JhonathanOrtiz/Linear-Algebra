```

vector = [1, 1]

def VectorPlot(vector, color, title=None):
  # Limites de la figura
  izda = min(-1, vector[0]-2)
  dcha = max(1, vector[0]+4)
  abajo = min(-1, vector[1]-2)
  arriba = max(1, vector[1]+4)

  # El metodo quiver pinta vectores, pero para que salgan de las
  # dimensiones correctas hay que usar los parámetros angles, scale y scale_units

  plt.quiver([vector[0]], [vector[1]], angles='xy', scale_units='xy', scale=1,color=color)

  # Pintamos lineas que pasan por el origen de coordenadas
  plt.axhline(0)
  plt.axvline(0)

  # Fijamos límites, etiquetas y títulos
  plt.xlim([izda, dcha])
  plt.ylim([abajo, arriba])
  plt.xlabel('x')
  plt.ylabel('y')
  plt.title(title)
  
  x1 = 1
  y1 = 1
  x2 = 1.5
  y2 = 0.5

VectorPlot(vector, 'blue')

plt.grid()
plt.show()
```

<h1>Un poco de teoría</h1>

Conocemos las operaciones elementales con vectores y la combinación lineal. El concepto de combinación lineal trae con el otro pequeño concepto que es *El subespacio generado.* Y esto son los vectores que podemos conseguir a partir de la combinación lineal de los vectores base, en éste caso el subespacio generado por los vectores i y j es el espacio porque podemos conseguir cualquier vector del espacio a partir de la combinación lineal.

```
#Construimos nuestros vectores base
i = j = [1, 0]

#Tomemos 2 escalares cualquiera
escalares = [(1, 4), (2, -1), (3, 2)]
colors = ['green', 'blue', 'red']

#Hacemos la combinación lineal
for index, par in enumerate(escalares):
  i = [par[0]*i[0], par[1]*i[1]]
  j = [par[0]*j[0], par[0]*j[1]]

  new_vector = [i[0]+j[0], i[1]+j[1]]
  VectorPlot(new_vector, colors[index], 'Vector generado ' + str(index))
  plt.grid()
  plt.show()
```

Hay momentos en los que los vectores unitarios dejan su lugar y pueden tomar cualquier valor prometo que en breve veremos porque puede pasar ésto, pero creeme que puede pasar, tenemos mala suerte y éstos dos vectores van a parar a la misma posición el subespacio generado por estos nuevos vectores se reduce a una linea y en ese caso decimos que esos dos vectores son linealmente dependientes, porque si los vectores tienen las mismas coordenadas podemos quitar cualquiera y el subespacio generado no se ve afectado. por el contrario si necesitamos ambos vectores para el subespacio generado los vectores son linealmente independientes y éste es el caso de los vectores anteriores.



Multiplicar una matriz y un vector siempre tiene este significado geometrico, el de saber adonde fue a parar un vector luego de una transformación. Tal como una función la matriz va a la izquierda... y toma un vector y te da un vector transformado.
