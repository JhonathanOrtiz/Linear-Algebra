```

vector = [1, 1]

def VectorPlot(vector, color, title=None):
  # Limites de la figura
  izda = min(-1, vector[0]-2)
  dcha = max(1, vector[0]+4)
  abajo = min(-1, vector[1]-2)
  arriba = max(1, vector[1]+4)

  # El metodo quiver pinta vectores, pero para que salgan de las
  # dimensiones correctas hay que usar los parmetros angles, scale y scale_units

  plt.quiver([vector[0]], [vector[1]], angles='xy', scale_units='xy', scale=1,color=color)

  # Pintamos lineas que pasan por el origen de coordenadas
  plt.axhline(0)
  plt.axvline(0)

  # Fijamos lmites, etiquetas y ttulos
  plt.xlim([izda, dcha])
  plt.ylim([abajo, arriba])
  plt.xlabel('x')
  plt.ylabel('y')
  plt.title(title)
  
  x1 = 1
  y1 = 1
  x2 = 1.5
  y2 = 0.5

VectorPlot(vector, 'blue')

plt.grid()
plt.show()
```

Conocemos las operaciones elementales con vectores y la combinacin lineal. El concepto de combinacin lineal trae con el otro pequeo concepto que es *El subespacio generado.* Y esto son los vectores que podemos conseguir a partir de la combinacin lineal de los vectores base, en ste caso el subespacio generado por los vectores i y j es el espacio porque podemos conseguir cualquier vector del espacio a partir de la combinacin lineal.

```
#Construimos nuestros vectores base
i = j = [1, 0]

#Tomemos 2 escalares cualquiera
escalares = [(1, 4), (2, -1), (3, 2)]
colors = ['green', 'blue', 'red']

#Hacemos la combinacin lineal
for index, par in enumerate(escalares):
  i = [par[0]*i[0], par[1]*i[1]]
  j = [par[0]*j[0], par[0]*j[1]]

  new_vector = [i[0]+j[0], i[1]+j[1]]
  VectorPlot(new_vector, colors[index], 'Vector generado ' + str(index))
  plt.grid()
  plt.show()
```
